## Funções de Manipulação de Arquivos

```python
def extrair_relatorios_csv(dicionario_registros, nome_arquivo):
    # Abre o arquivo CSV no modo de escrita ('w') para sobrescrever o arquivo existente
    with open(nome_arquivo, mode='w', newline='', encoding='utf-8') as arquivo_csv:

        # Percorre cada categoria (Despesas, Receitas, Investimentos) no dicionário
        for categoria, registros in dicionario_registros.items():
            # Escreve o nome da categoria como um título no arquivo CSV
            arquivo_csv.write(f'Categoria: {categoria}\n')

            # Cria um escritor CSV com as chaves do dicionário como cabeçalhos de coluna
            escritor_csv = csv.DictWriter(arquivo_csv, fieldnames=registros.keys(), delimiter=',')

            # Escreve o cabeçalho no arquivo CSV
            escritor_csv.writeheader()

            # Itera sobre os registros de acordo com o tamanho da lista 'id'
            for i in range(len(registros['id'])):
                # Cria um dicionário para representar cada linha de dados
                linha = {chave: registros[chave][i] for chave in registros}
                # Escreve a linha de dados no CSV
                escritor_csv.writerow(linha)

            # Adiciona uma linha em branco para separar categorias no arquivo CSV
            arquivo_csv.write('\n')

def extrair_relatorios_json(dicionario_registros, nome_arquivo):
    # Abre o arquivo JSON no modo de escrita ('w') para sobrescrever o arquivo existente
    with open(nome_arquivo, mode='w', encoding='utf-8') as arquivo_json:
        # Converte o dicionário em um formato JSON e escreve no arquivo
        # ensure_ascii=False mantém caracteres especiais
        # indent=4 adiciona indentação para facilitar a leitura do arquivo
        json.dump(dicionario_registros, arquivo_json, ensure_ascii=False, indent=4)
```

```python
import json

def ler_e_transformar_json(caminho_arquivo):
    # Abrir o arquivo JSON
    with open(caminho_arquivo, 'r', encoding='utf-8') as file:
        json_data = json.load(file)  # Carregar o JSON a partir do arquivo

    dicionario_registros = {}

    # Iterar pelas categorias no JSON
    for categoria, dados in json_data.items():
        dicionario_registros[categoria] = {}

        # Copiar os valores de 'id', 'Data', 'Valor' e 'Taxa' diretamente, se existirem
        for chave, valor in dados.items():
            dicionario_registros[categoria][chave] = valor

    return dicionario_registros


# Exemplo de uso:
caminho_arquivo = '/content/drive/My Drive/Cursos/Santander Coders 2024/LP II/Projeto Final/registros.json'

# Chamando a função para transformar o JSON
dicionario_transformado = ler_e_transformar_json(caminho_arquivo)

print(dicionario_transformado)
```

# Funções do Sistema

### Funções de Adicionar Valores

```python
def recebe_valida_valor():
    valor = 'inicia_while' # envia uma string aleatória para entrar no while
    while not isinstance(valor, float):
        valor = input('')
        try: # tenta converter pra float, caso dê erro mantém em string
            valor = float(valor)
        except:
            pass

        if not isinstance(valor, float):
            print('Valor inválido. Por favor, insira apenas valores numéricos.')
    return valor

def registra_id(transacao, dicionario_registros):
    if len(dicionario_registros[transacao]['id']) == 0 :
        dicionario_registros[transacao]['id'].append(0)
    else:
        dicionario_registros[transacao]['id'].append(dicionario_registros[transacao]['id'][-1]+1)
    return dicionario_registros


def registro_despesa(dicionario_registros):
    transacao = 'Despesas'
    print('Insira sua despesa abaixo. Caso deseje voltar ao menu, insira o valor 0.')
    print('Despesa (R$):')
    valor = recebe_valida_valor()

    if valor == 0:
        return dicionario_registros
    else:
        dicionario_registros = registra_id(transacao, dicionario_registros)

        valor_negativo = abs(valor)*(-1)
        dicionario_registros[transacao]['Data'].append(datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
        dicionario_registros[transacao]['Valor'].append(valor_negativo)
        return dicionario_registros


def registro_receita(dicionario_registros):
    transacao = 'Receitas'
    print('Insira sua receita abaixo. Caso deseje voltar ao menu, insira o valor 0.')
    print('Receita (R$):')
    valor = recebe_valida_valor()

    if valor == 0:
        return dicionario_registros
    else:
        dicionario_registros = registra_id(transacao, dicionario_registros)

        dicionario_registros[transacao]['Data'].append(datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
        dicionario_registros[transacao]['Valor'].append(valor)
        return dicionario_registros


def registro_investimento(dicionario_registros):
    transacao = 'Investimentos'
    print('Insira seu Investimento e sua Taxa abaixo. Caso deseje voltar ao menu, insira o valor 0.')
    print('Investimento (R$):')
    valor = recebe_valida_valor()
    print('Taxa (%):')
    taxa = recebe_valida_valor()

    if valor != 0 and taxa != 0:
        dicionario_registros = registra_id(transacao, dicionario_registros)

        dicionario_registros[transacao]['Data'].append(datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
        dicionario_registros[transacao]['Valor'].append(valor)
        dicionario_registros[transacao]['Taxa'].append(taxa)
        return dicionario_registros
    else:
        return dicionario_registros
```

### Funções de Editar Valores


```python
# Função para atualizr registro

def editar_registro(transacao, dicionario_registros):
    if transacao == 'Investimentos':
        print(f'id |         Data        |  Valor |  Taxa')
        for id, data, valor, taxa in zip(dicionario_registros[transacao]['id'], dicionario_registros[transacao]['Data'],
                                            dicionario_registros[transacao]['Valor'], dicionario_registros[transacao]['Taxa']):
            print('---------------------------------')
            print(f'{id}  | {data} | {valor} | {taxa}')
        print('---------------------------------\n')
    else:
        print(f'id |         Data        |  Valor')
        for id, data, valor  in zip(dicionario_registros[transacao]['id'], dicionario_registros[transacao]['Data'],
                                    dicionario_registros[transacao]['Valor']):
            print('---------------------------------')
            print(f'{id}  | {data} | {valor}')
        print('---------------------------------\n')

    escolha_id = 10
    while escolha_id != 'sair':
        escolha_id_str = input(f'Insira o identificador único do registro de {transacao} que deseja editar ou "sair" para voltar ao Menu:')
        try: # tenta converter pra float, caso dê erro mantém em string
            escolha_id = int(escolha_id_str)
        except:
            escolha_id = escolha_id_str

        if isinstance(escolha_id, int):
            if escolha_id in dicionario_registros[transacao]['id']:

                # Encontrar o índice correspondente ao ID na lista
                index = dicionario_registros[transacao]['id'].index(escolha_id)

                # Mostrar opções de atualização (exceto 'id') e adicionar a opção de sair
                while True:
                    print("Escolha o campo que deseja editar ou digite 'sair' para voltar:")
                    campos = [campo for campo in dicionario_registros[transacao].keys() if campo != 'id']  # Exclui o campo 'id'
                    for i, campo in enumerate(campos):
                        print(f"{i}. {campo}")
                    print(f"{len(campos)}. Sair")

                    # Pede para o usuário escolher um campo
                    escolha_campo_str = input("Digite o número do campo que deseja editar ou 'sair': ")
                    if escolha_campo_str == 'sair':
                        return dicionario_registros

                    try:
                        escolha_campo = int(escolha_campo_str)
                        if escolha_campo == len(campos):
                            print("Saindo sem atualizar o registro.")
                            return dicionario_registros
                        campo_selecionado = campos[escolha_campo]
                    except:
                        print("Opção inválida. Tente novamente.")
                        continue

                    # Solicita o novo valor para o campo selecionado
                    novo_valor = input(f"Digite o novo valor para {campo_selecionado}: ")

                    # Converte o tipo do valor se necessário
                    if campo_selecionado == 'Valor' or campo_selecionado == 'Taxa':  # Verifica se é numérico
                        try:
                            novo_valor = float(novo_valor)
                        except ValueError:
                            print(f"Valor inválido para {campo_selecionado}. Deve ser numérico.")
                            continue

                    # Atualiza o dicionário
                    dicionario_registros[transacao][campo_selecionado][index] = novo_valor

                print(f"Registro com ID {escolha_id} foi atualizado com sucesso.")

                return dicionario_registros
            else:
                print(f"ID {escolha_id} não encontrado. Por favor, insira um ID válido.")
        elif escolha_id == 'sair':
            pass
        else:
            print('Valor inválido. Por favor, insira um número inteiro ou "sair".')
    return dicionario_registros


def iniciar_editar(dicionario_registros):
    print(
        """
    Para prosseguir, selecione uma das opções abaixo:""")
    dicionario_seletor = {1: "Editar Despesas",
                          2: "Editar Receitas",
                          3: "Editar Investimentos",
                          0: "Voltar ao menu"}

    dicionario_funcoes = {
        1: "Despesas",
        2: "Receitas",
        3: "Investimentos",
        4: iniciar_sisfin,
        }

    escolha = 10  # Número aleatório diferente das opções do dicionario_seletor
    while escolha != 0:
        for chave, registro in dicionario_seletor.items(): # imprime o menu
            print(f"{chave}. {registro}")
        try: # validando se o valor que a pessoa inseriu é um valor que pode ser convertido para inteiro
            escolha = int(input("Sua opção: "))
        except: # se erro, insere um valor inválido em escolha para forçar o sistema a mostrar o print('Opção inválida...')
            escolha = 10
        if escolha == 0:
            print('Voltando ao menu...\n')
        elif escolha in dicionario_funcoes:
            transacao = dicionario_funcoes[escolha]
            editar_registro(transacao, dicionario_registros)
        else:
            print("Opção inválida. Por favor, escolha uma opção válida.\n")
```

### Função Deletar

```python
# Função para deletar registro
def deletar_registro(transacao , dicionario_registros):


    for chave, valor in dicionario_registros[transacao].items():
        print(f'{chave}: {valor}')

    while True:
        try:
            id = int(input(f'Insira o identificador único do registro de {transacao} para ser deletado: '))
            if id < 0:
                print("Valor inválido. Por favor, insira um número inteiro positivo.")
                continue
        except ValueError:
            print("Valor inválido. Por favor, insira um número inteiro.")
            continue

        if id in dicionario_registros[transacao]['id']:
            # Encontrar o índice correspondente ao ID na lista
            index = dicionario_registros[transacao]['id'].index(id)

            # Remover o item correspondente de todas as colunas
            del dicionario_registros[transacao]['id'][index]
            del dicionario_registros[transacao]['Data'][index]
            del dicionario_registros[transacao]['Valor'][index]
            if transacao == 'Investimentos':
                del dicionario_registros[transacao]['Taxa'][index]

            print(f"Registro com ID {id} foi deletado com sucesso.")

            # Exibir IDs restantes após a deleção
            ids_disponiveis = dicionario_registros[transacao]['id']

            if ids_disponiveis:
                print(f"IDs disponíveis após a deleção: {ids_disponiveis}")
                #print(f"Novo ID máximo: {max(ids_disponiveis)}")
            else:
                print("Não há mais registros disponíveis.")
            return dicionario_registros
        else:
            print(f"ID {id} não encontrado. Por favor, insira um ID válido.")

    return dicionario_registros

def iniciar_apagar(dicionario_registros):
    print(
        """
    Para prosseguir, selecione uma das opções abaixo:""")
    dicionario_seletor = {1: "Deletar Despesas",
                          2: "Deletar Receitas",
                          3: "Deletar Investimentos",
                          4: "Voltar ao menu",
                          0: "Sair do sistema"}

    dicionario_funcoes = {
        1: "Despesas",
        2: "Receitas",
        3: "Investimentos",
        4: iniciar_sisfin,
        }

    escolha = 10  # Número aleatório diferente das opções do dicionario_seletor
    while escolha != 0:
        for chave, registro in dicionario_seletor.items(): # imprime o menu
            print(f"{chave}. {registro}")
        try: # validando se o valor que a pessoa inseriu é um valor que pode ser convertido para inteiro
            escolha = int(input("Sua opção: "))
        except: # se erro, insere um valor inválido em escolha para forçar o sistema a mostrar o print('Opção inválida...')
            escolha = 10
        if escolha == 0:
            print('Saindo do sistema...')
            print('Logout realizado com sucesso!')
        elif escolha in dicionario_funcoes:
            if escolha == 1:
                transacao = "Despesas"
                deletar_registro(transacao , dicionario_registros)
                continue
            elif escolha == 2:
                transacao = "Receitas"
                deletar_registro(transacao , dicionario_registros)
                continue
            elif escolha == 3:
                transacao = "Investimentos"
                deletar_registro(transacao , dicionario_registros)
                continue
            elif escolha == 4:
                dicionario_funcoes[escolha]()
                #break necessario para o loop não continuar sem o usuário pedir
                break
            else:
                dicionario_funcoes[escolha](dicionario_registros)
                break
        else:
            print("Opção inválida. Por favor, escolha uma opção válida.\n")
```

#### Função Ler Registros

```python
### FUNÇÃO ler_registros(), conforme estrutura de dados Luiza

## utilizando kwargs, mas poderia ser (dicionario_registros, data=None, tipo=None, valor=None)
## sem especificação, somente (dicionario_registros), retorna todos os registros

def ler_registros(dicionario_registros, **argumentos_de_consulta):
    data = argumentos_de_consulta.get('data')
    tipo = argumentos_de_consulta.get('tipo')
    valor = argumentos_de_consulta.get('valor')

    registros_filtrados = []

    # Filtragem por tipo de transação ("Despesas", "Receitas" e "Investimentos")
    if tipo:
        if tipo in dicionario_registros:
            registros = {tipo: dicionario_registros[tipo]}
        else:
            print(f"Tipo '{tipo}' não encontrado.")
            return []
    else:
        registros = dicionario_registros

    # Filtragem por data e valor
    for transacao, valores in registros.items():
        for iteracao in range(len(valores['Data'])):
            data_registro = datetime.strptime(valores['Data'][iteracao], "%Y-%m-%d %H:%M:%S")
            if (data and data_registro.date() == datetime.strptime(data, "%d/%m/%Y").date()) or not data:
                if valor is None or ('Registro' in valores and valores['Registro'][iteracao] == valor) or ('Valor' in valores and valores['Valor'][iteracao] == valor):
                    registro = {'Tipo': transacao, 'Data': valores['Data'][iteracao]}
                    if 'Registro' in valores:
                        registro['Valor'] = valores['Registro'][iteracao]
                    elif 'Valor' in valores:
                        registro['Valor'] = valores['Valor'][iteracao]
                    if 'Taxa' in valores and iteracao < len(valores['Taxa']):
                        registro['Taxa'] = valores['Taxa'][iteracao]
                    registros_filtrados.append(registro)

    # Exibição dos registros filtrados em formato de tabela
    if registros_filtrados:
        print("\nRegistros Encontrados")
        header = ["Tipo", "Data", "Valor", "Taxa"]
        print(f"{header[0]:<15} {header[1]:<20} {header[2]:<15} {header[3]:<10}")
        print("=" * 60)

        for registro in registros_filtrados:
            tipo = registro['Tipo']
            data = registro['Data']
            valor = f"R$ {registro['Valor']:.2f}" if 'Valor' in registro else ""
            taxa = f"{registro['Taxa']}%" if 'Taxa' in registro else ""
            print(f"{tipo:<15} {data:<20} {valor:<15} {taxa:<10}")
    else:
        print("Nenhum registro encontrado.")
    return registros_filtrados
```

### Funções dentro do Submenu Ler Registro

```python
# Funções fora do Menu / Mais fácil conferir erros e realizar alterações

def consultar_por_data(dicionario_registros):
    data = input("Digite a data no formato DD/MM/AAAA: ")
    ler_registros(dicionario_registros, data=data)

def consultar_por_tipo(dicionario_registros):
    tipo = input("Digite o tipo (Despesas, Receitas, Investimentos): ")
    ler_registros(dicionario_registros, tipo=tipo)

def consultar_por_valor(dicionario_registros):
    valor = float(input("Digite o valor: "))
    ler_registros(dicionario_registros, valor=valor)

def consulta_geral(dicionario_registros):
    ler_registros(dicionario_registros)

def calcular_saldo_atual(dicionario_registros):
    saldo_despesas = sum(dicionario_registros['Despesas']['Valor'])
    saldo_receitas = sum(dicionario_registros['Receitas']['Valor'])
    saldo_atual = saldo_receitas + saldo_despesas
    print(f"{'='*40}")
    print(f"{'Categoria':<15} | {'Total (R$)':>20}")
    print(f"{'-'*40}")
    print(f"{'Despesas':<15} | {saldo_despesas:>20.2f}")
    print(f"{'Receitas':<15} | {saldo_receitas:>20.2f}")
    print(f"{'-'*40}")
    print(f"{'Saldo Atual':<15} | {saldo_atual:>20.2f}")
    print(f"{'='*40}")
    return saldo_atual
```

### Função calcular montante

```python
# Função  que calcula o montante calcular_montante()

# Função  que calcula o montante calcular_montante()

def calcular_montante(dicionario_registros):
    montantes = []
    total_investido = 0.0
    total_montante = 0.0
    data_atual = datetime.now()

    # Calcular montantes para cada investimento
    for i in range(len(dicionario_registros['Investimentos']['id'])):
        data_investimento = datetime.strptime(dicionario_registros['Investimentos']['Data'][i], '%Y-%m-%d %H:%M:%S')
        valor_investido = dicionario_registros['Investimentos']['Valor'][i]
        taxa_juros = dicionario_registros['Investimentos']['Taxa'][i] / 100.0

        dias_passados = (data_atual - data_investimento).days
        montante = valor_investido * (1 + taxa_juros)**dias_passados

        montantes.append({
            'Data do Investimento': data_investimento.strftime('%Y-%m-%d %H:%M:%S'),
            'Valor Investido': valor_investido,
            'Montante na Data Atual': round(montante, 2)
        })

        total_investido += valor_investido
        total_montante += montante

    # Formatação da tabela de saída
    print("\nRegistros de Montantes\n", "="*50)
    print(f"{'Data do Investimento':<25} | {'Valor Investido':<15} | {'Montante na Data Atual':<20}")
    print("="*50)

    for resultado in montantes:
        print(f"{resultado['Data do Investimento']:<25} | "
              f"R$ {resultado['Valor Investido']:<13.2f} | "
              f"R$ {resultado['Montante na Data Atual']:<18.2f}")

    print("="*50)
    print(f"{'Total':<25} | R$ {total_investido:<13.2f} | R$ {total_montante:<18.2f}")
    print("="*50)
    print("\n")
```
